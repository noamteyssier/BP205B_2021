# Import Modules

```python
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
import plotly.express as px
import plotly.io as pio
from tqdm import tqdm

import os
pio.templates.default = "plotly_white"
```

# Import Data

```python
def read_barcodes(dat, barcode_fn):
    with open(barcode_fn, "r") as f:
        arr = [
            line.split("-")[0] for line in f
        ]
        dat.obs['cell_barcode'] = arr
        dat.obs.set_index("cell_barcode", inplace=True)
    
    return dat

def read_genes(dat, gene_fn):
    gene_ids = []
    gene_names = []
    with open(gene_fn, "r") as f:
        for line in f:
            i, n = line.strip().split("\t")
            gene_ids.append(i)
            gene_names.append(n)
        
        dat.var['gene_id'] = gene_ids
        dat.var['gene_name'] = gene_names
        dat.var.set_index("gene_name", inplace=True)
    
    return dat

def load_anndata(matrix_path, barcodes_path, genes_path):
    adat = sc.read_mtx(matrix_path).T
    
    adat = read_barcodes(adat, barcodes_path)
    adat = read_genes(adat, genes_path)
    return adat

def build_ratio_frame(ms_frame):
    ratio_frame = []

    for idx, subframe in tqdm(ms_frame.groupby("Barcode")):
        
        sf = subframe.sort_values("nUMI", ascending=False)

        try:
            maj_b, min_b = sf.Multiseq[:2]
            maj_u, min_u = sf.nUMI[:2] + 1
        except ValueError:
            maj_b = sf.Multiseq.values[0]
            maj_u = sf.nUMI.values[0] + 1
            
            min_b = "nan"
            min_u = 1
            
        ratio = maj_u / min_u
        total = sf.nUMI.sum() + 2

        values = (idx, maj_b, min_b, maj_u, min_u, ratio, total)
        ratio_frame.append(values)


    ratio_frame = pd.DataFrame(
        ratio_frame, 
        columns = ['Barcode', 'maj_b', "min_b", "maj_u", "min_u", "ratio", "total"]
    )
    ratio_frame['log_ratio'] = np.log2(ratio_frame.ratio)
    ratio_frame['maj_f'] = ratio_frame.maj_u / ratio_frame.total
    return ratio_frame

def filter_ratio_frame(ratio_frame, min_lr = 2, min_total = 10):
    return ratio_frame[
        (ratio_frame.log_ratio >= min_lr) &
        (ratio_frame.total >= min_total)
    ]

def load_multiseq_mapping(mapping_path):
    ms_frame = pd.read_csv(mapping_path, sep="\t")
    ratio_frame = build_ratio_frame(ms_frame)
    ratio_frame = filter_ratio_frame(ratio_frame)
    return ratio_frame
    
def merge_multiseq(dat, ratio_frame):
    dat.obs = ratio_frame.\
        loc[:, ["Barcode", "maj_b"]].\
        rename(columns = {
            "Barcode" : "cell_barcode",
            "maj_b" : "MultiSeq"
        }).\
        set_index("cell_barcode").\
        merge(dat.obs, how = 'right', left_index=True, right_index=True)
    
    dat = dat[~dat.obs.MultiSeq.isna()]
    return dat

def merge_multiseq_labels(dat, ms_table):
    dat.obs = dat.obs.\
        reset_index().\
        merge(ms_table, how = 'left').\
        set_index("cell_barcode")
    return dat

def load_single_cell_dataset(input_dir, ms_table):
    
    matrix_path = os.path.join(input_dir, "matrix.mtx")
    barcodes_path = os.path.join(input_dir, "barcodes.tsv")
    genes_path = os.path.join(input_dir, "genes.tsv")
    mapping_path = os.path.join(input_dir, "ms_mapping.tab")
    
    adat = load_anndata(matrix_path, barcodes_path, genes_path)
    ratio_frame = load_multiseq_mapping(mapping_path)
    
    adat = merge_multiseq(adat, ratio_frame)
    adat = merge_multiseq_labels(adat, ms_table)
    adat.var_names_make_unique()
    
    return adat
    

ms_table = pd.read_csv("../data/MultiseqTable.tab", sep="\t")
path_aa = "../data/HCC_AA/filtered_gene_bc_matrices/GRCh38/"
path_mk = "../data/HCC_MK/filtered_gene_bc_matrices/GRCh38/"


adat_aa = load_single_cell_dataset(path_aa, ms_table)
adat_mk = load_single_cell_dataset(path_mk, ms_table)
adat = sc.concat([adat_aa, adat_mk])
```

# Filtering

```python
def calculate_qc(dat):
    
    # annotate the group of mitochondrial genes as 'mt'
    dat.var['mt'] = dat.var_names.str.startswith('MT-')  
    sc.pp.calculate_qc_metrics(
        dat, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True
    )
    dat = dat[:,(dat.var.n_cells_by_counts > 0) & (dat.var.total_counts > 0)]
    return dat
    
    
adat = calculate_qc(adat_aa)
adat = calculate_qc(adat_mk)

print(adat_aa.var.columns)
print(adat_aa.obs.columns)
```

```python
def filter_genes(dat, min_total = 1e1, min_cells = 1e1, min_mc = 1e-3):
    orig_size = dat.var.shape[0]
    
    dat = dat[:,
        (dat.var.total_counts >= min_total) &
        (dat.var.n_cells_by_counts >= min_cells) &
        (dat.var.mean_counts >= min_mc)
    ]
    
    new_size = dat.var.shape[0]
    print("Filtered : {} genes".format(orig_size - new_size))
    
    return dat

def norm(dat):
    sc.pp.normalize_total(dat, target_sum=1e6)
    sc.pp.log1p(dat)
    return dat

adat_aa = filter_genes(adat_aa, min_total = 1e2, min_mc = 1e-2)
adat_mk = filter_genes(adat_mk, min_total = 1e2, min_mc = 1e-2)

adat_aa = norm(adat_aa)
adat_mk = norm(adat_mk)
filt_adat = sc.concat([adat_mk, adat_aa])
```

```python
sc.pp.pca(filt_adat)
sc.pp.neighbors(filt_adat)
sc.tl.umap(filt_adat)
sc.pl.umap(filt_adat, color = ['Sample', 'Experiment', "total_counts"])
```

# Batch Normalization

## Harmony Integration

```python
sc.external.pp.harmony_integrate(filt_adat, "Experiment")
sc.pp.neighbors(filt_adat, use_rep="X_pca_harmony")
sc.tl.umap(filt_adat)
sc.pl.umap(filt_adat, color = ['Sample', 'Experiment', "total_counts"])
```

## Harmony Integration + Total Count Regression

```python
sc.pp.regress_out(filt_adat, keys = ['total_counts', "n_genes_by_counts"])
sc.pp.scale(filt_adat, max_value=10)
```

```python
sc.pp.pca(filt_adat)
sc.external.pp.harmony_integrate(filt_adat, "Experiment")
sc.pp.neighbors(filt_adat, use_rep="X_pca_harmony")
sc.tl.umap(filt_adat)
sc.pl.umap(filt_adat, color = ['Sample', 'Experiment', "total_counts", "n_genes_by_counts"])
```
